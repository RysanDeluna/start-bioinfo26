## Parte 2 — Manipulação de strings de DNA

```python
# Sequência de exemplo do gene TP53 (fragmento didático)
sequencia = 'ATGGAGGAGCCGCAGTCAGATCCTAGCGTTGAATCCTGAGGTCAATGAAATCTGTTCACTTGTGCCC'

print('Sequência:')
print(sequencia)
print(f'\nComprimento: {len(sequencia)} bases')
```

```python
# Slicing — extraindo regiões de interesse
primeiro_codon = sequencia[0:3]
print(f'Primeiro códon: {primeiro_codon}')  # Deve ser ATG (Met = início)

# Extraindo os primeiros 6 códons
for i in range(0, 18, 3):
    codon = sequencia[i:i+3]
    print(f'Posição {i}: {codon}')
```

```python
# Contagem de bases
contagem = {
    'A': sequencia.count('A'),
    'T': sequencia.count('T'),
    'G': sequencia.count('G'),
    'C': sequencia.count('C'),
}

print('Contagem de bases:')
for base, n in contagem.items():
    print(f'  {base}: {n}')

total = sum(contagem.values())
gc = (contagem['G'] + contagem['C']) / total
print(f'\nGC content: {gc:.2%}')
```

```python
# Transcrição DNA → RNA
rna = sequencia.replace('T', 'U')
print('mRNA transcrito:')
print(rna)
```

## Parte 3 — Funções reutilizáveis

```python
def calcular_gc(seq):
    """Calcula o GC content de uma sequência de DNA."""
    seq = seq.upper()
    gc = seq.count('G') + seq.count('C')
    return gc / len(seq)


def complemento_reverso(seq):
    """Retorna o complemento reverso de uma sequência de DNA."""
    mapa = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    seq = seq.upper()
    comp = ''.join(mapa[base] for base in seq)
    return comp[::-1]


def transcrever(seq):
    """Transcreve DNA para RNA mensageiro."""
    return seq.upper().replace('T', 'U')


# Testando as funções
seq_teste = 'ATGCGATCG'
print(f'Sequência:          {seq_teste}')
print(f'GC content:         {calcular_gc(seq_teste):.1%}')
print(f'Complemento reverso: {complemento_reverso(seq_teste)}')
print(f'mRNA:               {transcrever(seq_teste)}')
```

## Parte 4 — Parsing de formato FASTA

```python
# Simulando um arquivo FASTA como uma string multilinhas
fasta_texto = """>Â seq1|TP53_exon4|Homo_sapiens
ATGGAGGAGCCGCAGTCAGATCCTAGCG
TTGAATCCTGAGGTCAATGAAATCTGTT
>seq2|KRAS_exon2|Homo_sapiens
ATGACTGAATATAAACTTGTGGTAGTTG
GAGCTGGTGGCGTAGGCAAGAGTGCCAT
>seq3|EGFR_exon19|Homo_sapiens
ATGCGACCCTCCGGGACGGCCGGGGCAG
CCTCGGCTGGGCGGCGCGGCGGGCGGCC
"""

print('Arquivo FASTA (simulado):')
print(fasta_texto)
```

```python
def parsear_fasta(fasta_string):
    """
    Faz o parse de uma string no formato FASTA.
    Retorna um dicionário {header: sequencia}.
    """
    sequencias = {}
    header_atual = None
    seq_atual = []

    for linha in fasta_string.strip().split('\n'):
        linha = linha.strip()
        if linha.startswith('>'):
            # Salva a sequência anterior (se existir)
            if header_atual:
                sequencias[header_atual] = ''.join(seq_atual)
            # Inicia nova sequência
            header_atual = linha[1:]  # remove o '>'
            seq_atual = []
        elif linha:
            seq_atual.append(linha)

    # Salva a última sequência
    if header_atual:
        sequencias[header_atual] = ''.join(seq_atual)

    return sequencias


# Aplicando o parser
seqs = parsear_fasta(fasta_texto)

print(f'Total de sequências: {len(seqs)}\n')
for header, seq in seqs.items():
    print(f'Header: {header}')
    print(f'Sequência: {seq}')
    print(f'Tamanho: {len(seq)} bp | GC: {calcular_gc(seq):.1%}')
    print('---')
```

## Parte 5 — Entendendo erros comuns

```python
# IndexError — tente entender a mensagem antes de corrigir!
seq = 'ATG'
# Descomente a linha abaixo para ver o erro:
# print(seq[10])
```

```python
# KeyError — base não está no mapa de complemento
# Descomente para ver o erro:
# seq_com_n = 'ATGCGN'
# print(complemento_reverso(seq_com_n))
```

```python
# Como lidar com bases ambíguas usando .get() no dicionário
def complemento_reverso_robusto(seq):
    mapa = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    seq = seq.upper()
    comp = ''.join(mapa.get(base, 'N') for base in seq)  # 'N' para base desconhecida
    return comp[::-1]


print(complemento_reverso_robusto('ATGCGN'))  # Funciona! N vira N
```
